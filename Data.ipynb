```python
api = 'QV987QRTESUMF8WK'
```

```python
from alpha_vantage.timeseries import TimeSeries
```

```python
from alpha_vantage.techindicators import TechIndicators
```

```python
ts = TimeSeries(key=api,output_format = 'pandas')
```

```python
ts_IBM = ts.get_daily(symbol='IBM',outputsize='full')
```

```python
ti = TechIndicators(key = api,output_format ='pandas')
```

```python
ti_IBM = ti.get_sma(symbol='IBM',interval='daily',time_period = 10,series_type = 'close')
```

```python
import pandas as pd

# Assuming 'ts_IBM' is the data fetched and it contains the columns 'open', 'high', 'low', 'close'

# Extracting the relevant columns from ts_IBM (assuming ts_IBM[0] is a DataFrame)
df_IBM = ts_IBM[0]


df_IBM.reset_index(inplace=True)

# Display the first few rows to verify the data
print(df_IBM.head())
```

```python
print(df_IBM.columns)
print(df_IBM.head())
```

```python
df_IBM.rename(columns={
    '1. open': 'open',
    '2. high': 'high',
    '3. low': 'low',
    '4. close': 'close',
    '5. volume': 'volume'
}, inplace=True)
```

```python
df_IBM['average_price'] = df_IBM[['open', 'high', 'low', 'close']].mean(axis=1)
```

```python
# Calculate Price Range (High - Low)
df_IBM['price_range'] = df_IBM['high'] - df_IBM['low']

# Calculate Price Change (Close - Open)
df_IBM['price_change'] = df_IBM['close'] - df_IBM['open']

# Calculate Volatility (Standard Deviation of the daily price change)
df_IBM['volatility'] = df_IBM['price_change'].rolling(window=30).std()

# Displaying the first few rows with calculated metrics
print(df_IBM[['average_price', 'price_range', 'price_change', 'volatility']].head())
df_IBM['symbol'] = 'IBM'
```

```python
import pandas as pd
ti_IBM[0]

# Convert ti_IBM to a DataFrame
df_sma = ti_IBM[0]  # Extract the first element which contains the data
df_sma.reset_index(inplace=True)  # Reset the index to make 'date' a column

# Rename columns of df_sma for consistency (optional)
df_sma.rename(columns={'date': 'date', 'SMA': 'sma'}, inplace=True)

# Merge df_IBM with df_sma on the 'date' column
merged_df_IBM = pd.merge(df_IBM, df_sma[['date', 'sma']], on='date', how='left')

# Check the resulting merged DataFrame
print(merged_df_IBM.head())
```
