```python
#eXTRACTING DATE AND CLOSE COLUMN
df_1 = pd.read_csv("/content/Amazon.csv")

# Set 'Date' as index and extract only 'Close' column
df_1 = df_1.set_index("date")[["close"]]

# Display the first few rows
print(df_1.head())
import numpy as np
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Checking the Graph of Closing price
plt.figure(figsize=(10,5))
plt.plot(df_1["close"], label="Close Price")
plt.title("Stock Close Price Over Time")
plt.legend()
plt.show()

def adf_test(series):
    result = adfuller(series)
    print("ADF Statistic:", result[0])
    print("p-value:", result[1])
    if result[1] < 0.05:
        print("The series is stationary.")
    else:
        print("The series is NOT stationary.")

adf_test(df_1["close"])

df_1["Close_diff"] = df_1["close"].diff().dropna()
adf_test(df_1["Close_diff"].dropna())  # Check stationarity again
plot_acf(df_1["Close_diff"].dropna())  # Helps in choosing q
plot_pacf(df_1["Close_diff"].dropna())  # Helps in choosing p
plt.show()

# Define ARIMA model (Example: ARIMA(1,1,1))
model = ARIMA(df_1["close"], order=(1, 1, 1))  # Change (p, d, q) based on ACF/PACF
model_fit = model.fit()

# Print summary
print(model_fit.summary())

!pip install pytz
import pytz
# ... (rest of your code) ...

# Convert df_1.index to datetime if it's not already
df_1.index = pd.to_datetime(df_1.index)

# Explicitly specify timezone (e.g., UTC) when creating the date range
forecast_steps = 10  # Number of days to forecast
forecast = model_fit.forecast(steps=forecast_steps)

# Plot actual vs forecasted values
plt.figure(figsize=(10, 5))
plt.plot(df_1.index, df_1["close"], label="Actual Close Price")
plt.plot(pd.date_range(df_1.index[-1], periods=forecast_steps, freq="D", tz=pytz.utc),
         forecast, label="Forecast", color="red")  # Added tz=pytz.utc
plt.title("ARIMA Forecast on Close Price")
plt.legend()
plt.show()

from sklearn.metrics import mean_absolute_error, mean_squared_error

y_true = df_1['close'][-forecast_steps:]
y_pred = forecast[:len(y_true)]

mae = mean_absolute_error(y_true, y_pred)
rmse = np.sqrt(mean_squared_error(y_true, y_pred))
```
